```python
Aim : Write a program to implement K nearest neighbor algorithm. Randomly generate a data of 20 points having x and y values.
    Also assign positive values when  x or y component is below 26 and negative values to all other input data. 
    Perform classification by using value of K=1,2,3,4,5.
```

```python
from random import randint

data = {"pos":[], "neg": []}


for _ in range(20):

	x, y = randint(0, 50), randint(0, 50)

	if x<=25 and y<=50:
		data["pos"].append([x, y])
	else:
		data["neg"].append([x, y]) 


print("Positive Points")
for item in data["pos"]:
	print(item)
print()

print("Negative Points")
for item in data["neg"]:
	print(item)	
print('\n')

x = int(input("Enter x co-od: "))
y = int(input("Enter y co-od: "))

distances = []

for category in data:
	for pt in data[category]:
		x1, y1 = pt
		dist = ((x-x1)*2 + (y-y1)*2)*0.5
		distances.append([dist, pt])

distances.sort()

for k in range(1, 6):
	
	print("\nValue of K =", k)

	cntpos = cntneg = 0
	neighbours = distances[:k]

	for item in neighbours:
		print(*item)

	for item in neighbours:
		if item[1] in data["pos"]:
			cntpos += 1
		else:
			cntneg += 1	

	if cntpos == cntneg:
		if neighbours[0][1] in data["pos"]:
			print("Class - Positive.\n")
		else:
			print("Class - Negative.\n")
	elif cntpos>cntneg:
		print("Class - Positive.\n")
	else:
		print("Class - Negative.\n")
```
